**<u>Sales by employee</u>**

```python
import pandas as pd
import matplotlib.pyplot as plt

customers_df = pd.read_csv('/Users/maz/Library/Mobile Documents/com~apple~CloudDocs/Documents/Education/Digital Skills/Data /Level 4 Data Analytics/Week 7/Project/csv/Customers.csv')
sales_df = pd.read_csv('/Users/maz/Library/Mobile Documents/com~apple~CloudDocs/Documents/Education/Digital Skills/Data /Level 4 Data Analytics/Week 7/Project/csv/Sales.csv')
employees_df = pd.read_csv('/Users/maz/Library/Mobile Documents/com~apple~CloudDocs/Documents/Education/Digital Skills/Data /Level 4 Data Analytics/Week 7/Project/csv/Employees.csv')
products_df = pd.read_csv('/Users/maz/Library/Mobile Documents/com~apple~CloudDocs/Documents/Education/Digital Skills/Data /Level 4 Data Analytics/Week 7/Project/csv/Products.csv')
# Merge the Sales and Employees dataframes
sales_employees_df = pd.merge(sales_df, employees_df, left_on='SalesPersonID', right_on='EmployeeID')

# Group the data by the salesperson's name and sum the quantity column
sales_by_employee = sales_employees_df.groupby('FirstName')['Quantity'].sum()

# Create a pie chart
sales_by_employee.plot(kind='pie')

# Display the chart
plt.show()
```

This code creates a pie chart based on the first name of each employee that displays the total amount of sales made by each employee. In order to analyse the sales data and find trends and patterns, this can be helpful.

  

For instance, you could be able to identify the staff members who are most responsible for the company's overall sales, and you could use this knowledge to decide how to allocate resources or reward top-performing staff members. By examining historical trends in sales performance and applying this knowledge to project future sales, you may potentially utilise this data to make sales predictions.

  

The code also sums the quantity column, groups the sales data by employee first name, and creates a pie chart, which might be helpful for gauging each employee's overall sales performance. By looking at this information, you could spot sales performance trends or patterns that might not be obvious from the pie chart alone.
