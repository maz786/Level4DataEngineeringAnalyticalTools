**<u>Sales by Product</u>**

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

customers = pd.read_csv("https://raw.githubusercontent.com/maz786/Level4DataEngineeringAnalyticalTools/main/Final%20Project/Database/csv/Customers.csv")
sales = pd.read_csv("https://raw.githubusercontent.com/maz786/Level4DataEngineeringAnalyticalTools/main/Final%20Project/Database/csv/Sales.csv")
employees = pd.read_csv("https://raw.githubusercontent.com/maz786/Level4DataEngineeringAnalyticalTools/main/Final%20Project/Database/csv/Employees.csv")
products = pd.read_csv("https://raw.githubusercontent.com/maz786/Level4DataEngineeringAnalyticalTools/main/Final%20Project/Database/csv/Products.csv")

sales_products = pd.merge(sales, products, on='ProductID')

sales_by_product = sales_products.groupby('Name')['Quantity'].sum()
sales_by_product.plot(kind='bar')

plt.title('Sales by Product')
plt.xlabel('Product')
plt.ylabel('Quantity Sold')

plt.show()
```

Based on the information in the Sales and Products datasets, this code generates a bar chart that displays the total number of each product that has been sold. Depending on the situation and other data available, this chart may be useful for doing analysis, spotting trends, and even making sales projections.
