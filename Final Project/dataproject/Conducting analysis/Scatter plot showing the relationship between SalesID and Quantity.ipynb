**<u>Scatter plot showing the relationship between SalesID and Quantity</u>**

```python
import pandas as pd
import matplotlib.pyplot as plt
# Read the Sales database into a DataFrame
sales_df = pd.read_csv('/Users/maz/Library/Mobile Documents/com~apple~CloudDocs/Documents/Education/Digital Skills/Data /Level 4 Data Analytics/Week 7/Project/csv/Sales.csv')
# Create a scatter plot showing the relationship between SalesID and Quantity
plt.scatter(sales_df['SalesID'], sales_df['Quantity'])
plt.xlabel('SalesID')
plt.ylabel('Quantity')
plt.title('SalesID vs Quantity')
plt.show()
```
