**<u>Sales of Products with 'Mountain' or 'Frame' in the Name</u>**

```python
# First, we need to import the necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Next, we need to read in the CSV files and store them in dataframes
customers_df = pd.read_csv("/Users/maz/Library/Mobile Documents/com~apple~CloudDocs/Documents/Education/Digital Skills/Data /Level 4 Data Analytics/Week 7/Project/csv/Customers.csv")
sales_df = pd.read_csv("/Users/maz/Library/Mobile Documents/com~apple~CloudDocs/Documents/Education/Digital Skills/Data /Level 4 Data Analytics/Week 7/Project/csv/Sales.csv")
employees_df = pd.read_csv("/Users/maz/Library/Mobile Documents/com~apple~CloudDocs/Documents/Education/Digital Skills/Data /Level 4 Data Analytics/Week 7/Project/csv/Employees.csv")
products_df = pd.read_csv("/Users/maz/Library/Mobile Documents/com~apple~CloudDocs/Documents/Education/Digital Skills/Data /Level 4 Data Analytics/Week 7/Project/csv/Products.csv")

# Now we need to merge the Sales and Products dataframes on the ProductID field
sales_products_df = pd.merge(sales_df, products_df, on="ProductID")

# Now we can use the "Name" field in the Products dataframe to filter the data
# We'll use the "str.contains" method to check if the "Name" field contains either "mountain" or "frame"
sales_products_filtered_df = sales_products_df[sales_products_df["Name"].str.contains("mountain|frame", case=False)]

# Now we can use the "Quantity" field to sum up the total sales for each product
sales_products_filtered_df = sales_products_filtered_df.groupby("Name")["Quantity"].sum().reset_index()

# Now we can create a bar plot to visualize the total sales for each product
plt.bar(sales_products_filtered_df["Name"], sales_products_filtered_df["Quantity"])
plt.xlabel("Product Name")
plt.ylabel("Total Sales")
plt.title("Sales of Products with 'Mountain' or 'Frame' in the Name")
plt.show()
```
