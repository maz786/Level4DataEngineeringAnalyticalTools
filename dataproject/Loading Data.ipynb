```python
import pandas as pd
import matplotlib.pyplot as plt
# Read the Sales database into a DataFrame
sales_df = pd.read_csv('https://raw.githubusercontent.com/maz786/Level4DataEngineeringAnalyticalTools/main/Final%20Project/Database/csv/Sales.csv')
# Create a scatter plot showing the relationship between SalesID and Quantity
plt.scatter(sales_df['SalesID'], sales_df['Quantity'])
plt.xlabel('SalesID')
plt.ylabel('Quantity')
plt.title('SalesID vs Quantity')
plt.show()
```
